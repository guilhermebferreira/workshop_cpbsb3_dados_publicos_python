```python
#importando o pandas
import pandas as pd
```

```python
#carregando arquivo csv
df = pd.read_csv('2018.csv', sep=';', encoding='latin1', skiprows=1, decimal=',')
```

```python
#vamos visualizar os 5 primeiros registros do dataframe
df.head()
```

```python
#Verificando o tipo e o tamanho do dataframe
df.info()
```

```python
# Análise descritiva simples dos dados quantitativos
df.describe()
```

```python
#Análise descritiva dos campos textuais
df.describe(include='object')
```

```python
#Qual o valor total gasto pelos senadores em 2018
df['VALOR_REEMBOLSADO'].sum()
```

```python
#Qual o maior valor reembolsado
df['VALOR_REEMBOLSADO'].max()
```

```python
#Quais as informações completas do maior valor reembolsado
df[df['VALOR_REEMBOLSADO']==df['VALOR_REEMBOLSADO'].max()]
```

```python
#Quais as informações completas do menor valor reembolsado
df[df['VALOR_REEMBOLSADO']==df['VALOR_REEMBOLSADO'].min()]
```

```python
df.nlargest(5, 'VALOR_REEMBOLSADO')
```

```python
#Quantos pedidos cada senador fez
df['SENADOR'].value_counts()
```

```python
#Quantos reembolsos por tipo de despesa
df['TIPO_DESPESA'].value_counts()
```

```python
#Qual o valor total reembolsado pelo Sen. Fernando Collor
df[df['SENADOR']=='FERNANDO COLLOR']['VALOR_REEMBOLSADO'].sum()
```
