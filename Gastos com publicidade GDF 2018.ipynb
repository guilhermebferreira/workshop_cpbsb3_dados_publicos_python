# Análise dos gastos com publicidade do GDF em 2018

```python
import pandas as pd
```

```python
# carregando o arquivo e transformando em dataframe
df = pd.read_csv('pagamento-publicidade-2018-dados-abertos-anual.csv', encoding='latin1', sep=';',
                usecols=[2,3,4,5,7,9,16,24], decimal=',')
```

```python
df['Líquido'].max()
```

```python
#Alterando nome da coluna
df.rename(columns={df.columns[6]: 'liquido'}, inplace=True)

df.info()
```

```python
#Convertendo a coluna liquido para float
df['liquido'] = df['liquido'].astype(float)
```

```python
#Tranformando a coluna antes de converter


#removendo espaços em branco
df['liquido'] = df['liquido'].str.strip()

#removendo pontos de milhar
df['liquido'] = df['liquido'].str.replace('.', '')

#trocando virgula por pontor
df['liquido'] = df['liquido'].str.replace(',', '.')

#Convertendo a coluna liquido para float
df['liquido'] = df['liquido'].astype(float)

df.info()
```
